### Q. What are the locations with the fastest roller coasters? (minimum of 10 roller coasters at that location)

```python
df["Location"].value_counts()
```

```python
df.query('Location != "Other"')
```

```python
df.query('Location != "Other"').groupby("Location")["Speed_mph"].agg(['mean', "count"]).query("count >= 10").sort_values("mean")['mean'].plot(kind = "barh", figsize = (12, 5), 
                                                                                                                                              title = "Average Coast Speed by Location")
plt.xlabel("Average Coaster Speed", fontsize = 15)
```
